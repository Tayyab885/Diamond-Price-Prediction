### Data Preprocessing

```python
## Converting Categorical Data into numerical
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for feature in categorical_features:
    df[feature] = le.fit_transform(df[feature])
df.head()
```

```python
X = df.drop('Price(in US dollars)',axis=1)
X
```

```python
y = df['Price(in US dollars)']
y
```

### Spliting Data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=1)
```

```python
print(f"Train Data: {X_train.shape}")
print(f"Test Data: {X_test.shape}")
```

### Model Building

- ### Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt_model = DecisionTreeRegressor()
dt_model.fit(X_train,y_train)
```

### Checking Models Performance

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
dt_model.score(X_test,y_test)
```

```python
y_pred = dt_model.predict(X_test)
```

```python
mae = mean_absolute_error(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
dt_r2 = r2_score(y_test,y_pred)
rmse = np.sqrt(mse)
print(f"MAE is: {mae}.")
print(f"MSE is: {mse}.")
print(f"R2 is: {dt_r2}.")
print(f"RMSE is: {rmse}.")
```

- ### Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf_model = RandomForestRegressor()
rf_model.fit(X_train,y_train)
```

### Checking Models Performance

```python
rf_model.score(X_test,y_test)
```

```python
y_pred1 = rf_model.predict(X_test)
```

```python
mae = mean_absolute_error(y_test,y_pred1)
mse = mean_squared_error(y_test,y_pred1)
rf_r2 = r2_score(y_test,y_pred1)
rmse = np.sqrt(mse)
print(f"MAE is: {mae}.")
print(f"MSE is: {mse}.")
print(f"R2 is: {rf_r2}.")
print(f"RMSE is: {rmse}.")
```

### R2 Score Comparision

```python
print(f"Decision Tree Regressor {dt_r2 * 100}%")
print(f"Random Forest Regressor {rf_r2 * 100}%")
```
